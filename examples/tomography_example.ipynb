# Import necessary modules 

```python
%matplotlib inline
import numpy as np
import tifffile
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import matplotlib.pyplot as plt
from ganrec.utils import angles, nor_tomo
from ganrec.ganrec2 import GANtomo
```

# Read the data 

```python
prj = tifffile.imread('./test_data/tooth.tiff')
nang, px = prj.shape
ang = angles(nang)
prj = nor_tomo(prj)
```

# Reconstruct the data with the ganrec algorithm

```python
rec = GANtomo(prj, ang, iter_num=1000).recon
plt.imshow(rec)
plt.show()
tifffile.imwrite('./test_results/recon_tooth', rec)
```
