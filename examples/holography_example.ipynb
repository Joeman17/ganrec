# Load the necessary modules

```python
import time
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import tifffile
import matplotlib.pyplot as plt
from ganrec.utils import nor_phase
from ganrec.ganrec2 import GANphase
```

# Set the parameters and read the data

```python
energy = 10
z = 0.5
pv = 5e-7
iter_num = 1000
fname_data = './test_data/hologram_shepp.tiff'
data = tifffile.imread(fname_data)
data = nor_phase(data)
```

# Reconstruct the phase with GANrec and save the result

```python
abs, phase = GANphase(data, energy, z, pv, iter_num=iter_num).recon
plt.imshow(phase)
plt.show()
tifffile.imwrite('./test_results/phase_shepp', phase)
```
